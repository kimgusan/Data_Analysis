```python
from scipy.stats import iqr

# 전체 컬럼에 대한 이상치 확인
#'happiness_score', 'gdp_per_capita',
#'social_support', 'healthy_life_expectancy',
#'freedom_to_make_life_choices', 'generosity',
# 'perceptions_of_corruption'

# happiness_score
happiness_Q1 = happiness_df.describe().T.loc['happiness_score', '25%']
happiness_Q3 = happiness_df.describe().T.loc['happiness_score', '75%']

iqr_value = iqr(happiness_df.happiness_score)
print(iqr_value)

lower_bound = happiness_Q1 - 1.5 * iqr_value
upper_bound = happiness_Q3 + 1.5 * iqr_value
print(f'행복지수 정상치 범위: {lower_bound} ~ {upper_bound}')

# gdp_per_capita
print()
gdp_per_capita_Q1 = happiness_df.describe().T.loc['gdp_per_capita', '25%']
gdp_per_capita_Q3 = happiness_df.describe().T.loc['gdp_per_capita', '75%']
print(gdp_per_capita_Q1, gdp_per_capita_Q3)
iqr_value = iqr(happiness_df.gdp_per_capita)

lower_bound = gdp_per_capita_Q1 - 1.5 * iqr_value
upper_bound = gdp_per_capita_Q3 + 1.5 * iqr_value
print(f'gdp 성장률 정상치 범위: {lower_bound} ~ {upper_bound}')

# social_support
print()
social_support_Q1 = happiness_df.describe().T.loc['social_support', '25%']
social_support_Q3 = happiness_df.describe().T.loc['social_support', '75%'] 
print(social_support_Q1, social_support_Q3)
iqr_value = iqr(happiness_df.social_support)

lower_bound = social_support_Q1 - 1.5 * iqr_value
upper_bound = social_support_Q3 + 1.5 * iqr_value
print(f'사회지원 정상치 범위: {lower_bound} ~ {upper_bound}')

# healthy_life_expectancy
print()
healthy_life_expectancy_Q1 = happiness_df.describe().T.loc['healthy_life_expectancy', '25%']
healthy_life_expectancy_Q3 = happiness_df.describe().T.loc['healthy_life_expectancy', '75%']
print(healthy_life_expectancy_Q1, healthy_life_expectancy_Q3)

iqr_value = healthy_life_expectancy_Q3 - healthy_life_expectancy_Q1
# 아래 라이브러리를 사용한 경우 nan 이라는 문구 발생
# 결측치가 있기 떄문에 발생한 것으로 추정
# iqr_value = iqr(happiness_df.healthy_life_expectancy)
lower_bound = healthy_life_expectancy_Q1 - 1.5 * iqr_value
upper_bound = healthy_life_expectancy_Q3 + 1.5 * iqr_value
print(f'기대수명 정상치 범위: {lower_bound} ~ {upper_bound}')

# freedom_to_make_life_choices
print()
freedom_to_make_life_choices_Q1 = happiness_df.describe().T.loc['freedom_to_make_life_choices', '25%']
freedom_to_make_life_choices_Q3 = happiness_df.describe().T.loc['freedom_to_make_life_choices', '75%']
print(freedom_to_make_life_choices_Q1, freedom_to_make_life_choices_Q3)
iqr_value = iqr(happiness_df.gdp_per_capita)

lower_bound = freedom_to_make_life_choices_Q1 - 1.5 * iqr_value
upper_bound = freedom_to_make_life_choices_Q3 + 1.5 * iqr_value
print(f'자유도에 따른 삶 정상치 범위: {lower_bound} ~ {upper_bound}')

# generosity
print()
generosity_Q1 = happiness_df.describe().T.loc['generosity', '25%']
generosity_Q3 = happiness_df.describe().T.loc['generosity', '75%']
print(generosity_Q1, generosity_Q3)
iqr_value = iqr(happiness_df.generosity)

lower_bound = generosity_Q1 - 1.5 * iqr_value
upper_bound = generosity_Q3 + 1.5 * iqr_value
print(f'관대함에 따른 삶 정상치 범위: {lower_bound} ~ {upper_bound}')

# perceptions_of_corruption
print()
perceptions_of_corruption_Q1 = happiness_df.describe().T.loc['perceptions_of_corruption', '25%']
perceptions_of_corruption_Q3 = happiness_df.describe().T.loc['perceptions_of_corruption', '75%']
print(perceptions_of_corruption_Q1, perceptions_of_corruption_Q3)
iqr_value = iqr(happiness_df.perceptions_of_corruption)

lower_bound = perceptions_of_corruption_Q1 - 1.5 * iqr_value
upper_bound = perceptions_of_corruption_Q3 + 1.5 * iqr_value
print(f'국가부패에 따른 삶 정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
import numpy as np
# 결측치 있는 데이터를 제외한 데이터 프레임 생성

# 결측치 있는 데이터 확인 country = State of Palestine
country_out_of_value = happiness_df[happiness_df['healthy_life_expectancy'].isna()]
# display(country_out_of_value)


# 결측치 있는 데이터 삭제 후 새로운 데이터프레임 생성
country_out_of_value = happiness_df[~happiness_df['healthy_life_expectancy'].isna()]
display(country_out_of_value.info()) # index 137

# 해당 데이터프레임 나라 1개 기준으로 numpy 내적을 이용하여 증감치 확인 -> 각각의 전체 나라 확인
# 1. country, region 를 제외한 값을 array 로 변경
# 2. 데이터의 차원을 맞추기 위해 7개 data 확인
print(country_out_of_value.columns)
happiness_data_0_to_6 = country_out_of_value[['happiness_score']].iloc[0:6]
another_data_0_to_6 = country_out_of_value[['gdp_per_capita', 'social_support',\
                                            'healthy_life_expectancy', 'freedom_to_make_life_choices', \
                                            'generosity', 'perceptions_of_corruption']].iloc[0:6]
# print(happiness_data_0_to_6)
# print(another_data_0_to_6)



# 각 변량에 대한 차원 확인 
A = np.array(another_data_0_to_6.astype(np.float32))
k = np.array(happiness_data_0_to_6.astype(np.float32))
# print(A)
# print(k)
print(A.shape)
print(k.shape)

    
# Aw = k (선형대수를 이용한 종속변수의 증감량 확인) 
# 선형대수의 역행렬을 이용한 증감값을 계산 하였으나 정사각행렬일때만 계산이 가능하여 수치를 잘라서 계산,
# 피쳐가 많은 경우에는 최소 제곱법을 이용한 증감값을 사용하는 것이 일반적이라고 했음
hp_w = np.dot(np.linalg.inv(A), k)
hp_w_flatten =hp_w.flatten()


# DataFrame 형식을 이용하여 시각화
features = ['gdp_per_capita','social_support', 'healthy_life_expectancy','freedom_to_make_life_choices', 'generosity','perceptions_of_corruption']
result_df = pd.DataFrame({'hp_w_flatten':hp_w_flatten})
display(result_df)

result_df.index = features
display(result_df.T)


# 해당 증감치 중 종속변수에 가장 영향을 많이 끼치는 항목 확인
```

```python
# 이상치를 제거하여 출력해야하지 않을까?

import numpy as np
country_out_of_value = happiness_df[happiness_df['healthy_life_expectancy'].isna()]
country_out_of_value = happiness_df[~happiness_df['healthy_life_expectancy'].isna()]

result_df = pd.DataFrame()

features = ['gdp_per_capita','social_support', 'healthy_life_expectancy','freedom_to_make_life_choices', 'generosity','perceptions_of_corruption']
# result_df.index = features

for i in range(22):
    start_idx = i *6
    end_idx = start_idx + 6
    happiness_data_0_to_6 = country_out_of_value[['happiness_score']].iloc[start_idx:end_idx]
    another_data_0_to_6 = country_out_of_value[['gdp_per_capita', 'social_support',\
                                                'healthy_life_expectancy', 'freedom_to_make_life_choices', \
                                                'generosity', 'perceptions_of_corruption']].iloc[start_idx:end_idx]
    # 각 변량에 대한 차원 확인 및 타입 변경
    A = np.array(another_data_0_to_6.astype(np.float32))
    k = np.array(happiness_data_0_to_6.astype(np.float32))
    hp_w = np.dot(np.linalg.inv(A), k)
    hp_w_flatten =hp_w.flatten()
    hp_w_flatten = [float(item) for item in hp_w_flatten]
    
    column_name = f'range {start_idx}~{end_idx}'
    result_df[column_name] = hp_w_flatten

result_df.index = features
display(result_df.T)
```

```python
# 결측치 있는 데이터 확인 country = State of Palestine
country_in_of_value = happiness_df[happiness_df['healthy_life_expectancy'].isna()]
display(country_in_of_value)

# 결측치 제외한 데이터 확인
country_out_of_value = happiness_df[~happiness_df['healthy_life_expectancy'].isna()]
# display(country_out_of_value.columns)

# 결측치를 제외한 대륙별 피치간의 행복지수 확인
hp_mean_df = country_out_of_value.groupby('region').mean('all').reset_index()
display(hp_mean_df)
hp_mean_df_select = hp_mean_df[hp_mean_df['region'] == 'Middle East and North Africa']
display(hp_mean_df_select)

# # 최소값
hp_min_df = country_out_of_value.groupby('region').min('all').reset_index()
display(hp_min_df)
hp_min_df_select = hp_min_df[hp_min_df['region'] == 'Middle East and North Africa']
display(hp_min_df_select)

# 최대값
hp_max_df = country_out_of_value.groupby('region').max('all').reset_index()
display(hp_max_df)
hp_max_df_select = hp_max_df[hp_max_df['region'] == 'Middle East and North Africa']
display(hp_max_df_select)

# 행복지수 최빈값
hp_std_df = country_out_of_value.groupby('region')['happiness_score'].std().reset_index()
display(hp_std_df)
hp_std_df_select =hp_std_df[hp_std_df['region'] == 'Middle East and North Africa']
display(hp_std_df_select)
```

```python
all_results_df = pd.DataFrame()

for i in range(0, 137, 7):
    happiness_data = country_out_of_value[['happiness_score']].iloc[i:i+6]
    another_data = country_out_of_value[['gdp_per_capita', 'social_support', 'healthy_life_expectancy',
                                          'freedom_to_make_life_choices', 'generosity', 'perceptions_of_corruption']].iloc[i:i+6]
    
    A = np.array(another_data.astype(np.float32))
    k = np.array(happiness_data.astype(np.float32))
    
    # 최소 제곱 해 계산
    hp_w, _, _, _ = np.linalg.lstsq(A, k, rcond=None)
    hp_w = hp_w.flatten()
    
    # 결과 추가
    column_name = f"Range {i}-{i+5}"
    all_results_df[column_name] = hp_w

all_results_df.index = features
final_df = all_results_df.T
final_df

display(final_df.describe().T)
```
